```python
import requests
import os
import re

START_ID = 312310
END_ID = 312353
FOLDER = "Oswiadczenia_Pobrane_2026"
os.makedirs(FOLDER, exist_ok=True)

# Słowa, których NIE CHCESZ w nazwie pliku
CZARNA_LISTA = ["informacja-o-urzedzie", "uchwaly", "regulamin"]

s = requests.Session()
s.headers.update({"User-Agent": "Mozilla/5.0"})

for i in range(START_ID, END_ID + 1):
    url = f"https://bip.gdansk.pl/prawo-lokalne/radny,a,{i}"
    print(f"Sprawdzam ID {i}...")

    try:
        r = s.get(url, timeout=10)
        # Znajdujemy wszystkie linki do plików w kodzie
        pdf_links = re.findall(r'https://download\.cloudgdansk\.pl/[^"\'>\s]+\.pdf', r.text)

        # Pętla dla każdego znalezionego linku
        for link in set(pdf_links):
            nazwa_pliku = link.split('/')[-1].lower()

            # --- LOGIKA SPRAWDZANIA ŚMIECI (WŁAŚCIWE MIEJSCE) ---
            czy_zawiera_smieci = False
            for fraza in CZARNA_LISTA:
                if fraza in nazwa_pliku:
                    czy_zawiera_smieci = True

            # Jeśli znaleźliśmy śmieci, przerywamy pracę nad tym jednym plikiem i idziemy do następnego
            if czy_zawiera_smieci == True:
                print(f"  -> Pomijam śmieciowy plik: {nazwa_pliku}")
                continue
            # -----------------------------------------------------

            # Jeśli przeszło przez filtr, pobieramy
            print(f"  -> POBIERAM: {nazwa_pliku}")
            plik = s.get(link, timeout=20)
            with open(os.path.join(FOLDER, f"Radny_{i}_{nazwa_pliku}"), 'wb') as f:
                f.write(plik.content)

    except Exception as e:
        print(f"ID {i}: BŁĄD -> {e}")

print("ZAKOŃCZONO.")
```
